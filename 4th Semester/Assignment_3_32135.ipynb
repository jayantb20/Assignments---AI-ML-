```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv(r'Bwt.csv')
data
```

```python
#Testing the correlation between mother's age and baby's birth weight.
data['age'].corr(data['bwt'])
```

```python
#Generating the vectors from the dataset
X = data['age'].values.reshape(-1,1)
Y = data['bwt'].values.reshape(-1,1)
Z = data['weight'].values.reshape(-1,1)
temp1 = X
temp2 = Z
```

```python
#Spilting the data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train1, y_test1 = train_test_split(X, Y, test_size = 0.20 , random_state = 1)
X_train, X_test, Y_train1, y_test1
```

```python
#Training the data
reg1 = LinearRegression()
reg1.fit(X_train,Y_train1)
```

```python
#Testing and Predicting the bwt
prediction1 = reg1.predict(X_test)
prediction1 
```

```python
#Visualizing the motherâ€™s age by babyâ€™s birth weight and the line.
best_line1 = reg1.coef_ *temp1 + reg1.intercept_
print('Coefficients: ', reg1.coef_[0])
print('Intercept: ',reg1.intercept_)
plt.scatter(data['age'], data['bwt'], c= 'purple')
plt.plot(data['age'], best_line1 , c= 'yellow' ,linewidth = '2')
plt.title('Motherâ€™s age by babyâ€™s birth weight')
plt.xlabel('Motherâ€™s age')
plt.ylabel("Baby's weight (ounces)")
```

```python
#Checking Accuracy
from sklearn.metrics import r2_score
linreg_r2_score = r2_score(y_test1, prediction1)
reg1.score(X_train,Y_train1)
```

Conclusion:
From the above scatter plot it is clear that the mother's age and baby's birth weight are no where related and the correlation is poor.
